```python
from PIL import Image
import numpy as np 
import pandas as pd 
import torchvision
import os 
import json
import random
import ipyplot
```

```python
# Read settings
with open("settings.json", "r") as inFile:
    settings = json.load(inFile)

# Set random seed
np.random.seed(1984)
```

```python
# Parse list of patients 
pat_list = [pat for pat in os.listdir("Data/kaggle_3m") if "TCGA" in pat]

# Load image 
img = Image.open("Data/kaggle_3m/" + pat_list[10] + "/" + pat_list[10] + "_10.tif")
img.size
```

```python
# Parse list of patients 
pat_list = [pat for pat in os.listdir("Data/kaggle_3m") if "TCGA" in pat]

# Pick 9 random patients and plot 
pat_plot = random.sample(pat_list, 9)
pat_imgs = []
pat_imgs_mask = []
for pat in pat_plot:
    pat_img = Image.open("Data/kaggle_3m/" + pat + "/" + pat + "_10.tif")
    pat_mask =  Image.open("Data/kaggle_3m/" + pat + "/" + pat + "_10_mask.tif")
    pat_imgs.append(pat_img)
    pat_imgs_mask.append(pat_mask)
# Plot
ipyplot.plot_images(pat_imgs)
ipyplot.plot_images(pat_imgs_mask)
```
